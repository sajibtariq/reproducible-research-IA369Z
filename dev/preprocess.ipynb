```python
import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn import metrics
from sklearn.metrics import confusion_matrix
from sklearn.metrics import classification_report
from imblearn.over_sampling import SMOTE 
from sklearn import model_selection
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
from sklearn.naive_bayes import GaussianNB
from sklearn.svm import SVC
from sklearn.neural_network import MLPClassifier
from sklearn.model_selection import GridSearchCV
import warnings
warnings.simplefilter(action='ignore', category=Warning)
%matplotlib inline
```

```python
warnings.filterwarnings('ignore')
data = pd.read_csv('~/reproducible-research-IA369Z/data/Final_conventional.csv')

def class_l(x):

    if(x['MOS']>=1 and x['MOS']<2):
        return '1'  # poor
    if(x['MOS']>=2 and x['MOS']<3 ):
        return '2' # average
    if(x['MOS']>=3 and x['MOS']<5 ):
        return '3' # good
    
data=data.assign(mos=data.apply(class_l, axis=1)) 
print('----------------------------------------')
print()
print('Entire dataset including all the metrics')
print('----------------------------------------')
print()
print(data.head(3))
```

```python
print()
print()
feature_lab=data.loc[:,data.columns.isin(['RTT','Throughput','Packets','mos'])]#3
print('----------------------------------------')
print()
print('Dataset including only three feature and one target')
print('----------------------------------------')
print()
print(feature_lab.head(5))
```

```python

print()
print()
data=data.dropna() #drop entire row due to missing value
data['mos'].value_counts()

def class_l(x):

    if(x['mos']=='1'):
        return 'Poor'  # poor
    if(x['mos']=='2') :
        return 'Average' # average
    if(x['mos']=='3'):
        return 'Good' # good
    
data=data.assign(Mean_Opinion_Score=data.apply(class_l, axis=1))

sns.set(style="darkgrid")
p = sns.countplot(data=data, x = 'Mean_Opinion_Score')
p=p.get_figure()
print(p)
p.savefig('imbalanced_mos.pdf')

print('----------------------------------------')
print()
print('Distribution of MOS instances in dataset(imbalance)')
print('----------------------------------------')
print()
print(data.shape)
data['mos'].value_counts()
data
```

```python

print()
print()
feature=data.loc[:,data.columns.isin(['RTT','Throughput','Packets'])]#3
#X=X1.iloc[0:1725]
target=data.loc[:,data.columns.isin(['mos'])]#3
target.columns= ['mos']
#y=y1.iloc[0:1725]
#y=y1.iloc[0:1725]
print('----------------------------------------')
print()
print("'RTT','Throughput','Packets' select as feature and 'mos' select as target for further analysis") 
print('----------------------------------------')
print()
```

```python
print()
print()
feature.shape
target.shape
smote=SMOTE(random_state=42) # resample all class same
X, y=smote.fit_resample(feature,target)
print('----------------------------------------')
print()
print('Applied Synthetic Minority Over-sampling Technique (SMOTE) to balance the dataset')
print('----------------------------------------')
print()
print(X.shape, y.shape)

y_b=pd.DataFrame(y)
y_b.columns = ['mos']

def class_ll(x):

    if(x['mos']=='1'):
        return 'Poor'  # poor
    if(x['mos']=='2') :
        return 'Average' # average
    if(x['mos']=='3'):
        return 'Good' # good
    
y_b=y_b.assign(Mean_Opinion_Score=y_b.apply(class_ll, axis=1))

sns.set(style="darkgrid")
sn_plot=sns.countplot(data=y_b, x ='Mean_Opinion_Score')
sn_plot= sn_plot.get_figure()
print(sn_plot)
sn_plot.savefig("balanced_mos.pdf")

y_b['mos'].value_counts()
#y
#y=y.to_numpy()
#y
```

```python
print()
print()
print('----------------------------------------')
print()
print('Splited dataset into traning and testing sets at the ratio of 90% : 10 %')
print('----------------------------------------')
print()
X_train, X_test,y_train, y_test = train_test_split( X, y, test_size=0.1, random_state=42)


pd.DataFrame(X_train).to_csv("~/reproducible-research-IA369Z/data/X_train.csv", header=None, index=None)
pd.DataFrame(X_test).to_csv("~/reproducible-research-IA369Z/data/X_test.csv", header=None, index=None)
pd.DataFrame(y_train).to_csv("~/reproducible-research-IA369Z/data/y_train.csv", header=None, index=None)
pd.DataFrame(y_test).to_csv("~/reproducible-research-IA369Z/data/y_test.csv", header=None, index=None)
```
